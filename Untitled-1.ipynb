```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import colors
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
data = pd.read_csv("train.csv", sep=",")
```

```python
plt.figure(figsize=(12, 10))
num_cols_plot = ["Age", "Flight Distance", "Departure Delay in Minutes"]
for i, col in enumerate(num_cols_plot, 1):
    if col in data.columns:
        plt.subplot(3, 2, i)
        plt.hist(data[col].dropna(), bins=40)
        plt.title(col)
plt.tight_layout()
plt.show()
```

```python
print(data['Gender'].value_counts(), '\n')
print(data['Customer Type'].value_counts(), '\n')
print(data['Type of Travel'].value_counts(), '\n')
print(data['Class'].value_counts(), '\n')
print(data['satisfaction'].value_counts(), '\n')
```

```python
print('Number of missing data for Inflight wifi service', (data['Inflight wifi service'] == 0).sum())
print('Number of avaiable data for Inflight wifi service', (data['Inflight wifi service'] != 0).sum())
data = data[data['Inflight wifi service'] != 0]
data = data[data['Class'] != 'Eco Plus']

data['Gender'] = data['Gender'].map({'Female': 0, 'Male': 1})
data['Customer Type'] = data['Customer Type'].map({'Loyal Customer': 0, 'disloyal Customer': 1})
data['Type of Travel'] = data['Type of Travel'].map({'Business travel': 0, 'Personal Travel': 1})
data['Class'] = data['Class'].map({'Business': 0, 'Eco': 1})
data['satisfaction'] = data['satisfaction'].map({'neutral or dissatisfied': 0, 'satisfied': 1})

data = data.drop('id', axis=1)
satisfaction = data['satisfaction'].copy()
data = data.drop('satisfaction', axis=1)
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
```

```python
correlations = data.corr()
threshold = 0.5
high_correlations = correlations[(correlations >= threshold) | (correlations <= -1*threshold)]
sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
sns.heatmap(high_correlations, annot=True, cmap='coolwarm')
plt.show()
```

```python
data = data.drop(['Arrival Delay in Minutes'], axis=1)

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

pca = PCA(n_components=3, random_state=42)
pca.fit(data_scaled)
data_pca = pd.DataFrame(pca.transform(data_scaled), columns=(['col1', 'col2', 'col3']))
data_pca.describe().T
```

```python
elbow = KElbowVisualizer(KMeans(n_init=10), k=10)
elbow.fit(data_pca)
elbow.show()
```

```python
data_scaled = pd.DataFrame(data_scaled, columns = data.columns)

# Use k = 4 from Elbow
model = KMeans(n_clusters=4, random_state=42, n_init=10)
predictions = model.fit_predict(data_pca)
data_pca['Cluster'] = predictions
data['Cluster'] = predictions
data_scaled['Cluster'] = predictions

#Plotting the clusters
x = data_pca['col1']
y = data_pca['col2']
z = data_pca['col3']
cmap = colors.ListedColormap(["#682F2F", "#9E726F", "#D6B2B1", "#B9C0C9", "#9F8A78", "#F3AB60"])
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label='bla')
ax.scatter(x, y, z, s=40, c=data_pca['Cluster'], marker='o', cmap = cmap )
plt.show()
```

```python
#Plotting countplot of clusters
print('Relatively fairly distributed clusters.')
pal = ["#682F2F","#B9C0C9", "#9F8A78","#F3AB60"]
pl = sns.countplot(x = 'Cluster', data = data, hue = 'Cluster', palette = pal)
pl.set_title("Distribution Of Clusters")
plt.show()
```

```python
cluster = data.groupby(['Cluster']).mean().T
cluster.style.background_gradient(cmap='RdYlGn',axis=1)
```

```python
from sklearn.mixture import GaussianMixture

# ÐŸÐ¾Ð´Ð±Ð¾Ñ€ Ð¾Ð¿Ñ‚Ð¸Ð¼Ð°Ð»ÑŒÐ½Ð¾Ð³Ð¾ ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð° ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¾Ð² Ñ‡ÐµÑ€ÐµÐ· BIC/AIC
bic_scores = []
aic_scores = []
ks = range(2, 11)

for k in ks:
    gm = GaussianMixture(n_components=k, covariance_type='full', random_state=42)
    gm.fit(data_pca)
    bic_scores.append(gm.bic(data_pca))
    aic_scores.append(gm.aic(data_pca))
```

```python
plt.figure(figsize=(10, 5))
plt.plot(ks, bic_scores, marker='o', label='BIC')
plt.plot(ks, aic_scores, marker='o', label='AIC')
plt.xlabel('ÐšÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¾Ð²')
plt.ylabel('Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ BIC/AIC')
plt.title('Ð’Ñ‹Ð±Ð¾Ñ€ Ñ‡Ð¸ÑÐ»Ð° ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¾Ð² (EM)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
gm = GaussianMixture(n_components=4, covariance_type='full', random_state=42)
em_labels = gm.fit_predict(data_pca)

data_pca['EM_Cluster'] = em_labels
data['EM_Cluster'] = em_labels
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    data_pca['col1'], 
    data_pca['col2'], 
    data_pca['col3'], 
    c=em_labels, 
    cmap='tab10', 
    s=50
)

ax.set_title('EM Clusters (Gaussian Mixture)')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.show()
```

```python
sns.countplot(x='EM_Cluster', data=data, palette='tab10')
plt.title('Distribution of EM Clusters')
plt.show()
```

```python
cluster_profile = data.groupby('EM_Cluster').mean().T
cluster_profile.style.background_gradient(cmap='RdYlGn', axis=1)
```
